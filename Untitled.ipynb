```python
import numpy as np
import xarray as xr

import qnt.data as qndata
import qnt.output as qnout


def strategy_random_10_stocks(seed: int = 42, n_stocks: int = 10):
    # 1) Load stock data (OHLCV + is_liquid)
    data = qndata.stocks.load_data()

    # 2) Get a "tradable universe" snapshot from the last date (liquid stocks only)
    is_liquid_last = data.sel(field="is_liquid").isel(time=-1)
    assets_liquid = is_liquid_last.asset.values[is_liquid_last.values.astype(bool)]

    # Safety check: ensure we have enough liquid assets
    if len(assets_liquid) < n_stocks:
        raise ValueError(f"Not enough liquid assets ({len(assets_liquid)}) to pick {n_stocks} stocks.")

    # 3) Randomly pick 10 assets (fixed by seed so it’s reproducible)
    rng = np.random.default_rng(seed)
    chosen_assets = rng.choice(assets_liquid, size=n_stocks, replace=False)

    # 4) Build equal weights across all times for those assets
    times = data.time.values
    w = xr.DataArray(
        0.0,
        dims=("time", "asset"),
        coords={"time": times, "asset": data.asset.values},
        name="weights"
    )

    w.loc[dict(asset=chosen_assets)] = 1.0 / n_stocks

    # 5) Clean/normalize per Quantiacs rules (liquidity, NaNs, leverage, etc.)
    w = qnout.clean(data, w)
    return w


# Quantiacs typically expects you to expose "weights" (or call a backtest depending on template)
weights = strategy_random_10_stocks(seed=7, n_stocks=10)
print(weights)
```

```python
import numpy as np
import xarray as xr
import qnt.data as qndata
import qnt.output as qnout


def generate_random_equal_weight_portfolio(data, seed=7, n_stocks=10, rebalance_daily=True):
    """
    Picks n_stocks random liquid assets (based on last day liquidity) and returns:
      - chosen_assets (np.ndarray)
      - weights (xr.DataArray) with dims (time, asset), equal-weighted among chosen assets
    """
    close = data["close"]
    is_liquid = data["is_liquid"]

    # liquid assets on last day
    liquid_last = is_liquid.isel(time=-1).fillna(0).astype(bool)
    assets_liquid = close.asset.values[liquid_last.values]

    if len(assets_liquid) < n_stocks:
        raise ValueError(f"Not enough liquid assets ({len(assets_liquid)}) to pick {n_stocks}.")

    rng = np.random.default_rng(seed)
    chosen_assets = rng.choice(assets_liquid, size=n_stocks, replace=False)

    # build weights across all assets (Quantiacs expects full asset dimension)
    times = close.time.values
    all_assets = close.asset.values

    w = xr.DataArray(
        0.0,
        dims=("time", "asset"),
        coords={"time": times, "asset": all_assets},
        name="weights",
    )

    # equal-weight the chosen ones (0.1 each if n_stocks=10)
    w.loc[dict(asset=chosen_assets)] = 1.0 / n_stocks

    # Make it submission-safe (liquidity, NaNs, leverage normalization, etc.)
    w = qnout.clean(data, w)

    return chosen_assets, w


def compute_portfolio_mean_variance(data, chosen_assets, weights=None, annualize=252, use_log_returns=False):
    """
    Computes historical mean and variance of the portfolio using daily returns.
    If 'weights' is provided, uses the time-varying weights (after aligning).
    If 'weights' is None, assumes constant equal weights over chosen_assets.
    Returns daily + annualized mean/variance (+ annualized vol).
    """
    close = data["close"].sel(asset=chosen_assets)

    # returns
    if use_log_returns:
        rets = np.log(close / close.shift(time=1))
    else:
        rets = close / close.shift(time=1) - 1

    rets = rets.dropna("time")  # remove first NaN row

    # If time-varying weights supplied, align to rets time and chosen assets
    if weights is not None:
        w = weights.sel(asset=chosen_assets).sel(time=rets.time)
        # Ensure weights are normalized per day (just in case)
        w_sum = w.sum("asset")
        w = xr.where(w_sum != 0, w / w_sum, w)
        # portfolio daily return series: r_p(t) = sum_i w_i(t) * r_i(t)
        rp = (w * rets).sum("asset").values
        mu_p = float(np.nanmean(rp))
        var_p = float(np.nanvar(rp, ddof=1))
        # For reference, also compute cov-based variance using average weights
        # (optional; not returned)
    else:
        # constant equal weights
        R = rets.values  # (T, n)
        mu = np.nanmean(R, axis=0)          # (n,)
        Sigma = np.cov(R, rowvar=False)     # (n,n)
        n = len(chosen_assets)
        w = np.ones(n) / n
        mu_p = float(w @ mu)
        var_p = float(w @ Sigma @ w)

    mu_p_ann = mu_p * annualize
    var_p_ann = var_p * annualize
    vol_p_ann = float(np.sqrt(var_p_ann))

    return {
        "portfolio_mean_daily": mu_p,
        "portfolio_var_daily": var_p,
        "portfolio_mean_annual": mu_p_ann,
        "portfolio_var_annual": var_p_ann,
        "portfolio_vol_annual": vol_p_ann,
    }


# ---- Example usage ----
data = qndata.stocks.load_data()

chosen_assets, weights = generate_random_equal_weight_portfolio(data, seed=7, n_stocks=10)

stats_const = compute_portfolio_mean_variance(data, chosen_assets, weights=None)
print("Constant equal-weight stats:", stats_const)

stats_timevarying = compute_portfolio_mean_variance(data, chosen_assets, weights=weights)
print("Using time-varying weights stats:", stats_timevarying)
```

```python
import numpy as np
import xarray as xr
import qnt.data as qndata

def mean_and_variance_equal_weight_10(seed=7, n_stocks=10, annualize=252):
    data = qndata.stocks.load_data()

    close = data.sel(field="close")
    is_liquid = data.sel(field="is_liquid")

    # Pick 10 liquid assets based on last day
    is_liquid_last = is_liquid.isel(time=-1)
    assets_liquid = is_liquid_last.asset.values[is_liquid_last.values.astype(bool)]

    rng = np.random.default_rng(seed)
    chosen_assets = rng.choice(assets_liquid, size=n_stocks, replace=False)

    # Keep only those 10 assets
    close10 = close.sel(asset=chosen_assets)

    # Daily simple returns: r_t = close_t / close_{t-1} - 1
    rets10 = close10 / close10.shift(time=1) - 1
    rets10 = rets10.dropna("time")  # remove first NaN row

    # Convert to 2D array: (time, asset)
    R = rets10.values  # shape (T, 10)

    # Asset mean (daily) and covariance (daily)
    mu = np.nanmean(R, axis=0)              # shape (10,)
    Sigma = np.cov(R, rowvar=False)         # shape (10,10)

    # Equal weights
    w = np.ones(n_stocks) / n_stocks        # 0.1 each

    # Portfolio mean and variance (daily)
    mu_p = float(w @ mu)
    var_p = float(w @ Sigma @ w)

    # Annualize (common approximation)
    mu_p_ann = mu_p * annualize
    var_p_ann = var_p * annualize
    vol_p_ann = np.sqrt(var_p_ann)

    return {
        "chosen_assets": chosen_assets,
        "asset_mean_daily": mu,
        "asset_cov_daily": Sigma,
        "portfolio_mean_daily": mu_p,
        "portfolio_var_daily": var_p,
        "portfolio_mean_annual": mu_p_ann,
        "portfolio_var_annual": var_p_ann,
        "portfolio_vol_annual": vol_p_ann,
    }

res = mean_and_variance_equal_weight_10(seed=7, n_stocks=10)
print("Chosen assets:", res["chosen_assets"])
print("Portfolio mean (daily):", res["portfolio_mean_daily"])
print("Portfolio variance (daily):", res["portfolio_var_daily"])
print("Portfolio mean (annual):", res["portfolio_mean_annual"])
print("Portfolio vol (annual):", res["portfolio_vol_annual"])
```

```python
import xarray as xr

def get_field(data, field_name: str):
    """
    Returns DataArray(time, asset) for the requested field, handling both:
    - DataArray with 'field' dim
    - Dataset with variables
    """
    # Case A: DataArray with field dimension
    if isinstance(data, xr.DataArray) and ("field" in data.dims or "field" in data.coords):
        if field_name not in list(data.coords["field"].values):
            raise KeyError(f"Field '{field_name}' not found. Available fields: {list(data.coords['field'].values)}")
        return data.sel(field=field_name)

    # Case B: Dataset with variables
    if isinstance(data, xr.Dataset):
        if field_name in data.data_vars:
            return data[field_name]

        # Try common alternatives
        alternatives = {
            "close": ["close_adj", "adj_close", "last", "price", "Close"],
            "is_liquid": ["liquid", "isLiquid", "IsLiquid"],
        }
        for alt in alternatives.get(field_name, []):
            if alt in data.data_vars:
                return data[alt]

        raise KeyError(
            f"Variable '{field_name}' not found. Available variables: {list(data.data_vars)}"
        )

    raise TypeError(f"Unsupported data type: {type(data)}")
```
