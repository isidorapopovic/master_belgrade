```python
# ============================================================
# CELL 1: Imports
# ============================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cvxpy as cp
import xarray as xr

import qnt.data as qndata
```

```python
# ============================================================
# CELL 2: Random portfolio universe selection
# ============================================================
def load_random_stock_universe(n_assets=50, seed=42):
    """
    Load Quantiacs stock data and randomly select n_assets liquid stocks.

    Parameters
    ----------
    n_assets : int
        Number of assets to randomly select.
    seed : int
        Random seed for reproducibility.

    Returns
    -------
    data : xarray.DataArray
        Full Quantiacs dataset.
    close_sel : pandas.DataFrame
        Close prices of selected assets, indexed by time.
    rets_sel : pandas.DataFrame
        Daily simple returns of selected assets, indexed by time.
    selected_assets : list
        List of selected asset tickers.
    """

    # Load Quantiacs stock data
    data = qndata.stocks.load_data()

    # Extract close prices and liquidity flag
    close = data.sel(field="close")
    is_liquid = data.sel(field="is_liquid")

    # Keep only assets liquid on the latest date
    liquid_last = is_liquid.isel(time=-1).fillna(0).astype(bool)
    liquid_assets = close.asset.values[liquid_last.values]

    if len(liquid_assets) < n_assets:
        raise ValueError(
            f"Only {len(liquid_assets)} liquid assets available, "
            f"but n_assets={n_assets} was requested."
        )

    # Randomly select assets
    rng = np.random.default_rng(seed)
    selected_assets = rng.choice(liquid_assets, size=n_assets, replace=False).tolist()

    # Select close prices for chosen assets
    close_sel_xr = close.sel(asset=selected_assets)

    # Convert to pandas DataFrame: rows=time, columns=asset
    close_sel = close_sel_xr.to_pandas()

    # Clean prices
    close_sel = close_sel.ffill().dropna(how="all")

    # Compute simple daily returns
    rets_sel = close_sel.pct_change().fillna(0.0)

    return data, close_sel, rets_sel, selected_assets
```

```python
data, close_df, rets_df, assets = load_random_stock_universe(n_assets=50, seed=42)

print("Selected assets:")
print(assets[:10])   # first 10
print("\nClose prices shape:", close_df.shape)
print("Returns shape:", rets_df.shape)
close_df.head()
```
