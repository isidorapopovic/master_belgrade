```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
import qnt.data as qndata


# ============================================================
# 1) Load Quantiacs data + choose random liquid assets
# ============================================================
def load_quantiacs_universe(seed=7, n_assets=50):
    """
    Returns:
        data          : Quantiacs DataArray
        chosen_assets : np.array of selected asset tickers
    """
    data = qndata.stocks.load_data()   # dimensions usually: field, time, asset
    close = data.sel(field="close")
    is_liquid = data.sel(field="is_liquid")

    liquid_last = is_liquid.isel(time=-1).fillna(0).astype(bool)
    liquid_assets = close.asset.values[liquid_last.values]

    if len(liquid_assets) < n_assets:
        raise ValueError(
            f"Not enough liquid assets ({len(liquid_assets)}) to pick {n_assets}."
        )

    rng = np.random.default_rng(seed)
    chosen_assets = rng.choice(liquid_assets, size=n_assets, replace=False)

    return data, chosen_assets


# ============================================================
# 2) Return matrix R from close + rolling mu_t and Sigma_t
# ============================================================
def compute_returns_from_close(close, use_log_returns=False):
    """
    close: xarray DataArray with dims (time, asset)

    Returns:
        times : np.array of timestamps
        R     : np.ndarray of shape (T, n), daily returns
    """
    if use_log_returns:
        rets = np.log(close / close.shift(time=1))
    else:
        rets = close / close.shift(time=1) - 1

    rets = rets.dropna("time")
    times = rets.time.values
    R = rets.values.astype(float)

    return times, R


def compute_mu_sigma_series(
    data,
    chosen_assets,
    lookback=252,
    annualise=252,
    use_log_returns=False
):
    """
    Builds:
      - daily returns matrix R from close
      - rolling expected returns mu_t
      - rolling covariance matrices Sigma_t

    Returns:
        times        : np.array, shape (T,)
        mu_series    : np.ndarray, shape (T, n)
        Sigma_series : np.ndarray, shape (T, n, n)
        R            : np.ndarray, shape (T, n)
    """
    close = data.sel(field="close").sel(asset=chosen_assets)
    times, R = compute_returns_from_close(close, use_log_returns=use_log_returns)

    T, n = R.shape
    if T <= lookback + 1:
        raise ValueError(f"Not enough history: T={T}, need > {lookback+1}.")

    mu_series = np.zeros((T, n))
    Sigma_series = np.zeros((T, n, n))

    for t in range(T):
        start = max(0, t - lookback + 1)
        window = R[start:t+1, :]

        mu = np.nanmean(window, axis=0)

        if window.shape[0] >= 2:
            Sigma = np.cov(window, rowvar=False)
            if np.isscalar(Sigma):
                Sigma = np.array([[Sigma]])
        else:
            Sigma = np.eye(n) * 1e-6

        mu_series[t] = mu * annualise
        Sigma_series[t] = Sigma * annualise

    return times, mu_series, Sigma_series, R


# ============================================================
# 3) PSD enforcement for covariance
# ============================================================
def make_psd_matrix(Sigma, ridge=1e-3, eps=1e-8):
    """
    Returns a numerically safer PSD covariance matrix as numpy array.
    """
    S = 0.5 * (Sigma + Sigma.T)
    S = S + ridge * np.eye(S.shape[0]) + eps * np.eye(S.shape[0])
    return S


# ============================================================
# 4) Constraints and solver helpers
# ============================================================
def add_constraints(w, long_only=True, w_min=None, w_max=None, fully_invested=True):
    cons = []
    n = w.shape[0]

    if fully_invested:
        cons.append(cp.sum(w) == 1)

    if long_only:
        cons.append(w >= 0)

    if w_min is not None:
        cons.append(w >= w_min * np.ones(n))

    if w_max is not None:
        cons.append(w <= w_max * np.ones(n))

    return cons


def solve_problem(prob):
    """
    Try a few solvers in sequence.
    """
    solvers = [cp.OSQP, cp.SCS, cp.ECOS]
    last_error = None

    for solver in solvers:
        try:
            prob.solve(solver=solver, warm_start=True, verbose=False)
            if prob.status in ["optimal", "optimal_inaccurate"]:
                return True
        except Exception as e:
            last_error = e

    if last_error is not None:
        print("Warning: optimisation solver failed with:", repr(last_error))
    else:
        print("Warning: optimisation solver failed.")
    return False


def normalise_weights(w, fallback=None):
    """
    Clean and re-normalise weights.
    """
    if w is None:
        if fallback is None:
            raise ValueError("Weights are None and no fallback was provided.")
        w = fallback.copy()

    w = np.asarray(w, dtype=float)
    w = np.nan_to_num(w, nan=0.0, posinf=0.0, neginf=0.0)

    # project negatives away very lightly if numeric noise appears
    w[w < 0] = 0.0

    s = w.sum()
    if s <= 0:
        if fallback is None:
            w = np.ones_like(w) / len(w)
        else:
            w = np.asarray(fallback, dtype=float).copy()
            w = np.maximum(w, 0)
            w = w / w.sum()
    else:
        w = w / s

    return w


# ============================================================
# 5) One-step and two-step optimisation
# ============================================================
def solve_one_step(
    mu,
    Sigma,
    w_prev,
    gamma=5.0,
    lam=0.06,
    w_max=0.10,
    ridge=1e-3
):
    """
    Single-step optimisation:
        max mu'w - gamma * w'Sigma w - lam * ||w - w_prev||_1
    """
    n = len(mu)
    w = cp.Variable(n)

    S = make_psd_matrix(Sigma, ridge=ridge)
    turnover_penalty = cp.norm1(w - w_prev)

    objective = cp.Maximize(
        mu @ w - gamma * cp.quad_form(w, S) - lam * turnover_penalty
    )
    constraints = add_constraints(w, long_only=True, w_max=w_max, fully_invested=True)

    prob = cp.Problem(objective, constraints)
    ok = solve_problem(prob)

    if not ok:
        return normalise_weights(w_prev)

    return normalise_weights(w.value, fallback=w_prev)


def solve_two_step(
    mu,
    Sigma,
    w_prev,
    gamma=5.0,
    rho=25.0,
    lam=0.06,
    w_max=0.10,
    ridge=1e-3
):
    """
    Two-step optimisation:

    Step 1:
        target portfolio w*
        max mu'w* - gamma * w*'Sigma w*

    Step 2:
        trade towards target with turnover penalty
        min (rho/2)||w - w*||^2 + lam ||w - w_prev||_1
    """
    n = len(mu)
    S = make_psd_matrix(Sigma, ridge=ridge)

    # Step 1: target portfolio
    w_star = cp.Variable(n)
    objective1 = cp.Maximize(mu @ w_star - gamma * cp.quad_form(w_star, S))
    constraints1 = add_constraints(w_star, long_only=True, w_max=w_max, fully_invested=True)

    prob1 = cp.Problem(objective1, constraints1)
    ok1 = solve_problem(prob1)

    if not ok1:
        w_star_val = normalise_weights(w_prev)
    else:
        w_star_val = normalise_weights(w_star.value, fallback=w_prev)

    # Step 2: controlled move towards target
    w = cp.Variable(n)
    tracking_term = (rho / 2.0) * cp.sum_squares(w - w_star_val)
    turnover_penalty = lam * cp.norm1(w - w_prev)

    objective2 = cp.Minimise(tracking_term + turnover_penalty)
    constraints2 = add_constraints(w, long_only=True, w_max=w_max, fully_invested=True)

    prob2 = cp.Problem(objective2, constraints2)
    ok2 = solve_problem(prob2)

    if not ok2:
        w_val = normalise_weights(w_prev)
    else:
        w_val = normalise_weights(w.value, fallback=w_prev)

    return w_star_val, w_val


# ============================================================
# 6) Run optimiser through time
# ============================================================
def run_strategy(
    mu_series,
    Sigma_series,
    method="one",
    gamma=5.0,
    lam=0.06,
    rho=25.0,
    w_max=0.10,
    ridge=1e-3,
    w0=None
):
    """
    Returns:
        W      : realised weights over time, shape (T, n)
        W_star : target weights for two-step, NaN for one-step
    """
    T, n = mu_series.shape

    if w0 is None:
        w_prev = np.ones(n) / n
    else:
        w_prev = normalise_weights(w0)

    W = np.zeros((T, n))
    W_star = np.full((T, n), np.nan)

    for t in range(T):
        mu_t = mu_series[t]
        Sigma_t = Sigma_series[t]

        if method == "one":
            w_t = solve_one_step(
                mu_t, Sigma_t, w_prev,
                gamma=gamma, lam=lam, w_max=w_max, ridge=ridge
            )
        elif method == "two":
            w_star_t, w_t = solve_two_step(
                mu_t, Sigma_t, w_prev,
                gamma=gamma, rho=rho, lam=lam, w_max=w_max, ridge=ridge
            )
            W_star[t] = w_star_t
        else:
            raise ValueError("method must be 'one' or 'two'")

        W[t] = w_t
        w_prev = w_t

    return W, W_star


# ============================================================
# 7) Helpers: turnover, realised returns, equity, stats
# ============================================================
def turnover(W):
    """
    Per-period turnover:
        sum_i |w_t - w_{t-1}|
    Returns shape (T-1,)
    """
    return np.sum(np.abs(np.diff(W, axis=0)), axis=1)


def realised_portfolio_returns(R, W):
    """
    Causal realised returns:
      R[t] is the return from t-1 -> t
      so use W[t-1]

    Returns shape (T-1,)
    """
    return np.sum(R[1:] * W[:-1], axis=1)


def equity_curve(rp, start=1.0):
    """
    rp: portfolio returns, shape (T-1,)
    """
    return start * np.cumprod(1.0 + rp)


def max_drawdown(eq):
    peak = np.maximum.accumulate(eq)
    dd = eq / peak - 1.0
    return np.min(dd)


def summary_stats(rp, periods_per_year=252):
    """
    Returns a dictionary of performance statistics.
    """
    rp = np.asarray(rp, dtype=float)
    rp = rp[np.isfinite(rp)]

    if len(rp) == 0:
        return {
            "n_obs": 0,
            "mean_daily": np.nan,
            "vol_daily": np.nan,
            "ann_return": np.nan,
            "ann_vol": np.nan,
            "sharpe": np.nan,
            "max_drawdown": np.nan,
            "cum_return": np.nan
        }

    mean_daily = np.mean(rp)
    vol_daily = np.std(rp, ddof=1) if len(rp) > 1 else 0.0

    ann_return = mean_daily * periods_per_year
    ann_vol = vol_daily * np.sqrt(periods_per_year)

    sharpe = ann_return / ann_vol if ann_vol > 0 else np.nan

    eq = equity_curve(rp)
    mdd = max_drawdown(eq)
    cum_ret = eq[-1] - 1.0

    return {
        "n_obs": len(rp),
        "mean_daily": mean_daily,
        "vol_daily": vol_daily,
        "ann_return": ann_return,
        "ann_vol": ann_vol,
        "sharpe": sharpe,
        "max_drawdown": mdd,
        "cum_return": cum_ret
    }


def print_stats(name, stats):
    print(f"\n{name}")
    print("-" * len(name))
    for k, v in stats.items():
        if isinstance(v, (float, np.floating)):
            print(f"{k:15s}: {v: .6f}")
        else:
            print(f"{k:15s}: {v}")


# ============================================================
# 8) Main experiment
# ============================================================
if __name__ == "__main__":
    # -------------------------
    # Parameters
    # -------------------------
    seed = 7
    n_assets = 50
    lookback = 252

    gamma = 5.0
    lam = 0.06
    rho = 25.0
    w_max = 0.10        # max 10% per asset
    ridge = 1e-3
    use_log_returns = False

    # -------------------------
    # Load data
    # -------------------------
    data, chosen_assets = load_quantiacs_universe(seed=seed, n_assets=n_assets)
    print("Chosen assets (first 10):", chosen_assets[:10], "... total:", len(chosen_assets))

    # -------------------------
    # Build returns matrix R and rolling moments
    # -------------------------
    times, mu_series, Sigma_series, R_daily = compute_mu_sigma_series(
        data=data,
        chosen_assets=chosen_assets,
        lookback=lookback,
        annualise=252,
        use_log_returns=use_log_returns
    )

    print("R_daily shape:", R_daily.shape)
    print("mu_series shape:", mu_series.shape)
    print("Sigma_series shape:", Sigma_series.shape)

    # -------------------------
    # Run one-step and two-step strategies
    # -------------------------
    W_one, _ = run_strategy(
        mu_series, Sigma_series,
        method="one",
        gamma=gamma, lam=lam, rho=rho, w_max=w_max, ridge=ridge
    )

    W_two, W_star = run_strategy(
        mu_series, Sigma_series,
        method="two",
        gamma=gamma, lam=lam, rho=rho, w_max=w_max, ridge=ridge
    )

    # -------------------------
    # Helpers: turnover
    # -------------------------
    to_one = turnover(W_one)
    to_two = turnover(W_two)

    print("\nAverage turnover")
    print("----------------")
    print("One-step:", np.mean(to_one))
    print("Two-step:", np.mean(to_two))

    # -------------------------
    # Realised returns and equity
    # -------------------------
    rp_one = realised_portfolio_returns(R_daily, W_one)
    rp_two = realised_portfolio_returns(R_daily, W_two)

    eq_one = equity_curve(rp_one, start=1.0)
    eq_two = equity_curve(rp_two, start=1.0)

    # -------------------------
    # Stats
    # -------------------------
    stats_one = summary_stats(rp_one, periods_per_year=252)
    stats_two = summary_stats(rp_two, periods_per_year=252)

    print_stats("One-step stats", stats_one)
    print_stats("Two-step stats", stats_two)

    # ============================================================
    # 9) Plots
    # ============================================================

    # A) Turnover
    plt.figure(figsize=(10, 4))
    plt.plot(to_one, label="One-step turnover")
    plt.plot(to_two, label="Two-step turnover")
    plt.title("Turnover per rebalance: sum |Δw|")
    plt.xlabel("t")
    plt.ylabel("sum |Δw|")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # B) Weight heatmaps
    plt.figure(figsize=(10, 5))
    plt.imshow(W_one.T, aspect="auto")
    plt.title("One-step weights heatmap (assets x time)")
    plt.xlabel("t")
    plt.ylabel("asset")
    plt.colorbar(label="weight")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.imshow(W_two.T, aspect="auto")
    plt.title("Two-step weights heatmap (assets x time)")
    plt.xlabel("t")
    plt.ylabel("asset")
    plt.colorbar(label="weight")
    plt.tight_layout()
    plt.show()

    # C) A few asset weights over time
    k = min(5, len(chosen_assets))

    plt.figure(figsize=(10, 4))
    for i in range(k):
        plt.plot(W_one[:, i], label=f"{chosen_assets[i]}")
    plt.title("One-step: sample asset weights")
    plt.xlabel("t")
    plt.ylabel("weight")
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 4))
    for i in range(k):
        plt.plot(W_two[:, i], label=f"{chosen_assets[i]}")
    plt.title("Two-step: sample asset weights")
    plt.xlabel("t")
    plt.ylabel("weight")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # D) Tracking distance for two-step
    valid = np.isfinite(W_star).all(axis=1)
    dist = np.linalg.norm(W_two[valid] - W_star[valid], axis=1)

    plt.figure(figsize=(10, 4))
    plt.plot(dist)
    plt.title("Two-step tracking distance ||w_t - w*_t||_2")
    plt.xlabel("t")
    plt.ylabel("L2 distance")
    plt.tight_layout()
    plt.show()

    # E) Realised equity curves
    plt.figure(figsize=(10, 4))
    plt.plot(eq_one, label="One-step equity")
    plt.plot(eq_two, label="Two-step equity")
    plt.title("Realised equity curves (weights shifted by 1 day)")
    plt.xlabel("t")
    plt.ylabel("equity")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
