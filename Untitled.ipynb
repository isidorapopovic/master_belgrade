```python
# ============================================================
# CELL 2: Random portfolio universe selection
# ============================================================
def load_random_stock_universe(n_assets=50, seed=42):
    """
    Load Quantiacs stock data and randomly select n_assets liquid stocks.

    Parameters
    ----------
    n_assets : int
        Number of assets to randomly select.
    seed : int
        Random seed for reproducibility.

    Returns
    -------
    data : xarray.DataArray
        Full Quantiacs dataset.
    close_sel : pandas.DataFrame
        Close prices of selected assets, indexed by time.
    rets_sel : pandas.DataFrame
        Daily simple returns of selected assets, indexed by time.
    selected_assets : list
        List of selected asset tickers.
    """

    # Load Quantiacs stock data
    data = qndata.stocks.load_data()

    # Extract close prices and liquidity flag
    close = data.sel(field="close")
    is_liquid = data.sel(field="is_liquid")

    # Keep only assets liquid on the latest date
    liquid_last = is_liquid.isel(time=-1).fillna(0).astype(bool)
    liquid_assets = close.asset.values[liquid_last.values]

    if len(liquid_assets) < n_assets:
        raise ValueError(
            f"Only {len(liquid_assets)} liquid assets available, "
            f"but n_assets={n_assets} was requested."
        )

    # Randomly select assets
    rng = np.random.default_rng(seed)
    selected_assets = rng.choice(liquid_assets, size=n_assets, replace=False).tolist()

    # Select close prices for chosen assets
    close_sel_xr = close.sel(asset=selected_assets)

    # Convert to pandas DataFrame: rows=time, columns=asset
    close_sel = close_sel_xr.to_pandas()

    # Clean prices
    close_sel = close_sel.ffill().dropna(how="all")

    # Compute simple daily returns
    rets_sel = close_sel.pct_change().fillna(0.0)

    return data, close_sel, rets_sel, selected_assets
```

```python
# ============================================================
# CELL 3: Random feasible portfolio weights with max-weight cap
# ============================================================
def random_feasible_weights(n_assets, w_max=0.10, seed=None):
    """
    Generate random long-only portfolio weights such that:
      - weights sum to 1
      - each weight is between 0 and w_max

    Parameters
    ----------
    n_assets : int
        Number of assets in the portfolio.
    w_max : float
        Maximum allowed weight per asset.
    seed : int or None
        Random seed for reproducibility.

    Returns
    -------
    w : numpy.ndarray
        Feasible weight vector of shape (n_assets,).
    """

    if w_max <= 0:
        raise ValueError("w_max must be positive.")

    if w_max > 1:
        raise ValueError("w_max cannot be greater than 1.")

    if n_assets * w_max < 1:
        raise ValueError(
            f"No feasible long-only fully invested portfolio exists because "
            f"n_assets * w_max = {n_assets * w_max:.3f} < 1."
        )

    rng = np.random.default_rng(seed)

    # Start with zero weights
    w = np.zeros(n_assets)

    # Remaining total weight to allocate
    remaining = 1.0
    available = np.ones(n_assets, dtype=bool)

    # Randomly allocate until fully invested
    while remaining > 1e-12:
        idx = np.where(available)[0]

        if len(idx) == 0:
            raise RuntimeError("No assets left to allocate weight to, but portfolio is not full.")

        # Pick one asset randomly
        i = rng.choice(idx)

        # Maximum we can still assign to this asset
        cap_left = w_max - w[i]

        # To keep future feasibility, compute minimum that must remain for others
        other_idx = idx[idx != i]
        max_possible_others = len(other_idx) * w_max

        # Minimum amount we must assign here so that the rest can still be filled feasibly
        min_assign = max(0.0, remaining - max_possible_others)

        # Maximum feasible assignment here
        max_assign = min(cap_left, remaining)

        if max_assign < min_assign - 1e-12:
            raise RuntimeError("Feasibility issue during random weight generation.")

        # Randomly assign within feasible range
        add = rng.uniform(min_assign, max_assign)
        w[i] += add
        remaining -= add

        # Mark asset unavailable if it reached cap
        if w[i] >= w_max - 1e-12:
            available[i] = False

        # Safety against tiny numerical leftovers
        if remaining < 1e-12:
            remaining = 0.0

    # Final cleanup for numerical precision
    w = np.clip(w, 0, w_max)
    w = w / w.sum()

    return w
```

```python
# ============================================================
# CELL: Convex portfolio optimization functions
# ============================================================
import numpy as np
import cvxpy as cp


def project_to_psd(S, eps=1e-8):
    """
    Project a symmetric matrix onto the PSD cone.
    Useful because sample covariance matrices may be numerically indefinite.
    """
    S = 0.5 * (S + S.T)
    eigvals, eigvecs = np.linalg.eigh(S)
    eigvals_clipped = np.clip(eigvals, eps, None)
    return eigvecs @ np.diag(eigvals_clipped) @ eigvecs.T


def validate_inputs(mu, Sigma, w_prev, w_max):
    """
    Basic shape and feasibility checks.
    """
    mu = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)
    w_prev = np.asarray(w_prev).flatten()

    n = len(mu)

    if Sigma.shape != (n, n):
        raise ValueError(f"Sigma must have shape ({n}, {n}), got {Sigma.shape}.")
    if len(w_prev) != n:
        raise ValueError(f"w_prev must have length {n}, got {len(w_prev)}.")
    if w_max is not None and n * w_max < 1:
        raise ValueError(
            f"Infeasible constraints: n_assets * w_max = {n * w_max:.3f} < 1."
        )

    return mu, Sigma, w_prev


def build_constraints(w, n, w_max=None, long_only=True, fully_invested=True):
    """
    Standard portfolio constraints.
    """
    constraints = []

    if fully_invested:
        constraints.append(cp.sum(w) == 1)

    if long_only:
        constraints.append(w >= 0)

    if w_max is not None:
        constraints.append(w <= w_max)

    return constraints


def solve_one_step_portfolio(mu, Sigma, w_prev, gamma=1.0, tc_lambda=0.01,
                             w_max=0.10, long_only=True, fully_invested=True,
                             solver=cp.OSQP, verbose=False):
    """
    One-step optimization:
        maximize mu^T w - gamma * w^T Sigma w - tc_lambda * ||w - w_prev||_1

    Returns
    -------
    w_opt : np.ndarray
        Optimal portfolio weights
    info : dict
        Diagnostics
    """
    mu, Sigma, w_prev = validate_inputs(mu, Sigma, w_prev, w_max)
    n = len(mu)
    Sigma_psd = project_to_psd(Sigma)

    w = cp.Variable(n)
    constraints = build_constraints(
        w, n, w_max=w_max, long_only=long_only, fully_invested=fully_invested
    )

    objective = cp.Maximize(
        mu @ w
        - gamma * cp.quad_form(w, Sigma_psd)
        - tc_lambda * cp.norm1(w - w_prev)
    )

    problem = cp.Problem(objective, constraints)
    problem.solve(solver=solver, verbose=verbose, warm_start=True)

    if w.value is None:
        raise RuntimeError("One-step optimization failed. No solution returned.")

    w_opt = np.asarray(w.value).flatten()

    info = {
        "status": problem.status,
        "objective_value": problem.value,
        "turnover": float(np.sum(np.abs(w_opt - w_prev))),
        "expected_return": float(mu @ w_opt),
        "risk": float(w_opt @ Sigma_psd @ w_opt),
    }

    return w_opt, info


def solve_target_portfolio(mu, Sigma, gamma=1.0, w_max=0.10,
                           long_only=True, fully_invested=True,
                           solver=cp.OSQP, verbose=False):
    """
    Step 1 of two-step optimization:
        maximize mu^T w - gamma * w^T Sigma w

    Returns
    -------
    w_target : np.ndarray
        Target portfolio
    info : dict
        Diagnostics
    """
    mu = np.asarray(mu).flatten()
    Sigma = np.asarray(Sigma)
    n = len(mu)

    if Sigma.shape != (n, n):
        raise ValueError(f"Sigma must have shape ({n}, {n}), got {Sigma.shape}.")
    if w_max is not None and n * w_max < 1:
        raise ValueError(
            f"Infeasible constraints: n_assets * w_max = {n * w_max:.3f} < 1."
        )

    Sigma_psd = project_to_psd(Sigma)

    w = cp.Variable(n)
    constraints = build_constraints(
        w, n, w_max=w_max, long_only=long_only, fully_invested=fully_invested
    )

    objective = cp.Maximize(
        mu @ w - gamma * cp.quad_form(w, Sigma_psd)
    )

    problem = cp.Problem(objective, constraints)
    problem.solve(solver=solver, verbose=verbose, warm_start=True)

    if w.value is None:
        raise RuntimeError("Target portfolio optimization failed. No solution returned.")

    w_target = np.asarray(w.value).flatten()

    info = {
        "status": problem.status,
        "objective_value": problem.value,
        "expected_return": float(mu @ w_target),
        "risk": float(w_target @ Sigma_psd @ w_target),
    }

    return w_target, info


def trade_toward_target(w_target, w_prev, tc_lambda=0.01, w_max=0.10,
                        long_only=True, fully_invested=True,
                        solver=cp.OSQP, verbose=False):
    """
    Step 2 of two-step optimization:
        minimize 0.5 * ||w - w_target||_2^2 + tc_lambda * ||w - w_prev||_1

    Returns
    -------
    w_trade : np.ndarray
        Actual traded portfolio after partial adjustment
    info : dict
        Diagnostics
    """
    w_target = np.asarray(w_target).flatten()
    w_prev = np.asarray(w_prev).flatten()

    n = len(w_target)
    if len(w_prev) != n:
        raise ValueError(f"w_prev must have length {n}, got {len(w_prev)}.")
    if w_max is not None and n * w_max < 1:
        raise ValueError(
            f"Infeasible constraints: n_assets * w_max = {n * w_max:.3f} < 1."
        )

    w = cp.Variable(n)
    constraints = build_constraints(
        w, n, w_max=w_max, long_only=long_only, fully_invested=fully_invested
    )

    objective = cp.Minimize(
        0.5 * cp.sum_squares(w - w_target)
        + tc_lambda * cp.norm1(w - w_prev)
    )

    problem = cp.Problem(objective, constraints)
    problem.solve(solver=solver, verbose=verbose, warm_start=True)

    if w.value is None:
        raise RuntimeError("Trade-toward-target optimization failed. No solution returned.")

    w_trade = np.asarray(w.value).flatten()

    info = {
        "status": problem.status,
        "objective_value": problem.value,
        "turnover": float(np.sum(np.abs(w_trade - w_prev))),
        "distance_to_target_l1": float(np.sum(np.abs(w_trade - w_target))),
        "distance_to_target_l2": float(np.linalg.norm(w_trade - w_target)),
    }

    return w_trade, info


def solve_two_step_portfolio(mu, Sigma, w_prev, gamma=1.0, tc_lambda=0.01,
                             w_max=0.10, long_only=True, fully_invested=True,
                             solver=cp.OSQP, verbose=False):
    """
    Full two-step optimization wrapper:
      1) compute target portfolio
      2) trade toward target with turnover penalty

    Returns
    -------
    w_final : np.ndarray
        Final traded portfolio
    result : dict
        Contains target, final weights, and diagnostics
    """
    w_target, info_target = solve_target_portfolio(
        mu=mu,
        Sigma=Sigma,
        gamma=gamma,
        w_max=w_max,
        long_only=long_only,
        fully_invested=fully_invested,
        solver=solver,
        verbose=verbose,
    )

    w_final, info_trade = trade_toward_target(
        w_target=w_target,
        w_prev=w_prev,
        tc_lambda=tc_lambda,
        w_max=w_max,
        long_only=long_only,
        fully_invested=fully_invested,
        solver=solver,
        verbose=verbose,
    )

    result = {
        "w_target": w_target,
        "w_final": w_final,
        "target_info": info_target,
        "trade_info": info_trade,
    }

    return w_final, result
```

```python
# ============================================================
# CELL: Backtest one-step vs two-step
# ============================================================
def estimate_mu_sigma(returns_window, mu_lookback=20, cov_lookback=252):
    """
    Estimate expected returns and covariance from a rolling window.

    Parameters
    ----------
    returns_window : pd.DataFrame
        Historical returns up to current date.
    mu_lookback : int
        Lookback window for expected returns.
    cov_lookback : int
        Lookback window for covariance.

    Returns
    -------
    mu_t : np.ndarray
        Estimated expected returns (annualized).
    Sigma_t : np.ndarray
        Estimated covariance matrix (annualized).
    """
    mu_hist = returns_window.tail(mu_lookback)
    cov_hist = returns_window.tail(cov_lookback)

    mu_t = mu_hist.mean().values * 252.0
    Sigma_t = cov_hist.cov().values * 252.0

    Sigma_t = project_to_psd(Sigma_t)
    return mu_t, Sigma_t


def backtest_strategy(
    rets_df,
    method="one_step",
    gamma=1.0,
    tc_lambda=0.01,
    w_max=0.10,
    rebalance_every=21,
    mu_lookback=20,
    cov_lookback=252,
    start_index=None,
    initial_weights=None,
):
    """
    Backtest one-step or two-step convex portfolio optimization.

    Parameters
    ----------
    rets_df : pd.DataFrame
        Asset returns, index=time, columns=assets.
    method : str
        'one_step' or 'two_step'
    gamma : float
        Risk aversion.
    tc_lambda : float
        Turnover penalty.
    w_max : float
        Maximum weight per asset.
    rebalance_every : int
        Rebalance frequency in days.
    mu_lookback : int
        Lookback for expected returns.
    cov_lookback : int
        Lookback for covariance.
    start_index : int or None
        First day to start backtest. If None, uses max(mu_lookback, cov_lookback).
    initial_weights : np.ndarray or None
        Starting portfolio. If None, equal-weight.

    Returns
    -------
    result : dict
        Dictionary with backtest outputs.
    """
    assets = rets_df.columns.tolist()
    dates = rets_df.index
    n_assets = len(assets)
    T = len(rets_df)

    if start_index is None:
        start_index = max(mu_lookback, cov_lookback)

    if initial_weights is None:
        w_prev = np.ones(n_assets) / n_assets
    else:
        w_prev = np.asarray(initial_weights).flatten().copy()

    # Storage
    weights_hist = np.zeros((T, n_assets))
    port_rets = np.zeros(T)
    turnover = np.zeros(T)
    expected_ret_hist = np.full(T, np.nan)
    risk_hist = np.full(T, np.nan)

    for t in range(T):
        # Use currently held weights to earn today's return
        r_t = rets_df.iloc[t].values
        port_rets[t] = float(w_prev @ r_t)
        weights_hist[t, :] = w_prev

        # Rebalance only after enough history is available
        if t < start_index:
            continue

        if (t - start_index) % rebalance_every != 0:
            continue

        returns_window = rets_df.iloc[:t]
        mu_t, Sigma_t = estimate_mu_sigma(
            returns_window,
            mu_lookback=mu_lookback,
            cov_lookback=cov_lookback
        )

        if method == "one_step":
            w_new, info = solve_one_step_portfolio(
                mu=mu_t,
                Sigma=Sigma_t,
                w_prev=w_prev,
                gamma=gamma,
                tc_lambda=tc_lambda,
                w_max=w_max
            )

        elif method == "two_step":
            w_new, result = solve_two_step_portfolio(
                mu=mu_t,
                Sigma=Sigma_t,
                w_prev=w_prev,
                gamma=gamma,
                tc_lambda=tc_lambda,
                w_max=w_max
            )
            info = {
                "turnover": result["trade_info"]["turnover"],
                "expected_return": float(mu_t @ w_new),
                "risk": float(w_new @ Sigma_t @ w_new),
            }

        else:
            raise ValueError("method must be 'one_step' or 'two_step'")

        turnover[t] = np.sum(np.abs(w_new - w_prev))
        expected_ret_hist[t] = info["expected_return"]
        risk_hist[t] = info["risk"]

        # Rebalance at end of day t, new weights used from day t+1 onward
        w_prev = w_new.copy()

    equity = np.cumprod(1.0 + port_rets)

    result = {
        "method": method,
        "dates": dates,
        "assets": assets,
        "weights": pd.DataFrame(weights_hist, index=dates, columns=assets),
        "portfolio_returns": pd.Series(port_rets, index=dates, name=f"{method}_returns"),
        "equity": pd.Series(equity, index=dates, name=f"{method}_equity"),
        "turnover": pd.Series(turnover, index=dates, name=f"{method}_turnover"),
        "expected_return": pd.Series(expected_ret_hist, index=dates, name=f"{method}_expected_return"),
        "risk": pd.Series(risk_hist, index=dates, name=f"{method}_risk"),
        "parameters": {
            "gamma": gamma,
            "tc_lambda": tc_lambda,
            "w_max": w_max,
            "rebalance_every": rebalance_every,
            "mu_lookback": mu_lookback,
            "cov_lookback": cov_lookback,
            "start_index": start_index,
        }
    }

    return result
```

```python
# ============================================================
# CELL: Performance summary
# ============================================================
def max_drawdown(equity_curve):
    running_max = equity_curve.cummax()
    drawdown = equity_curve / running_max - 1.0
    return drawdown.min()

def annualized_sharpe(port_rets, periods_per_year=252):
    mu = port_rets.mean()
    sigma = port_rets.std()
    if sigma == 0:
        return np.nan
    return np.sqrt(periods_per_year) * mu / sigma

summary = pd.DataFrame({
    "One-Step": {
        "Final Equity": res_one["equity"].iloc[-1],
        "Sharpe": annualized_sharpe(res_one["portfolio_returns"]),
        "Max Drawdown": max_drawdown(res_one["equity"]),
        "Average Turnover": res_one["turnover"].mean(),
    },
    "Two-Step": {
        "Final Equity": res_two["equity"].iloc[-1],
        "Sharpe": annualized_sharpe(res_two["portfolio_returns"]),
        "Max Drawdown": max_drawdown(res_two["equity"]),
        "Average Turnover": res_two["turnover"].mean(),
    }
}).T

summary
```
